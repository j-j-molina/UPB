# **Data preparation**

```python
data['no_of_adults'] = data['no_of_adults'].astype(int)
data['no_of_children'] = data['no_of_children'].astype(int)
data['no_of_weekend_nights'] = data['no_of_weekend_nights'].astype(int)
data['no_of_week_nights'] = data['no_of_week_nights'].astype(int)
data['type_of_meal_plan'] = data['type_of_meal_plan'].astype(object)
data['required_car_parking_space'] = data['required_car_parking_space'].astype(int)
data['room_type_reserved'] = data['room_type_reserved'].astype(object)
data['lead_time'] = data['lead_time'].astype(int)
data['arrival_year'] = data['arrival_year'].astype(object)
data['arrival_month'] = data['arrival_month'].astype(int)
data['arrival_date'] = data['arrival_date'].astype(int)
data['market_segment_type'] = data['market_segment_type'].astype(object)
data['repeated_guest'] = data['repeated_guest'].astype(int)
data['no_of_previous_cancellations'] = data['no_of_previous_cancellations'].astype(int)
data['no_of_previous_bookings_not_canceled'] = data['no_of_previous_bookings_not_canceled'].astype(int)
data['avg_price_per_room'] = data['avg_price_per_room'].astype(float)
data['no_of_special_requests'] = data['no_of_special_requests'].astype(int)
data['booking_status'] = data['booking_status'].astype(object)
```

```python
data.info()
```

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()
varNum=['no_of_adults','no_of_children','no_of_weekend_nights','no_of_week_nights','required_car_parking_space',
        'lead_time','arrival_month','arrival_date','no_of_previous_cancellations','no_of_previous_bookings_not_canceled','avg_price_per_room','no_of_special_requests']
min_max_scaler.fit(data[varNum])
data[varNum]= min_max_scaler.transform(data[varNum])
data.head()
```

```python
data = pd.get_dummies(data, columns=['type_of_meal_plan','room_type_reserved','market_segment_type'], drop_first=False, dtype=int)
data = pd.get_dummies(data, columns=['arrival_year'], drop_first=True, dtype=int)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
data["booking_status"]=labelencoder.fit_transform(data["booking_status"])
data.head()
```

```python
data.info()
```

# **Data division**

```python
from sklearn.model_selection import train_test_split
X = data.drop("booking_status", axis = 1)
Y = data['booking_status']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, stratify=Y)
medidas= pd.DataFrame(index=['mse','rmse','mae','mape','max'])
```

# **Decision tree ensamble**

```python
from sklearn.tree import DecisionTreeClassifier
model_dt = DecisionTreeClassifier(criterion='gini', min_samples_leaf=30, max_depth=15)
model_dt.fit(X_train, Y_train)

from sklearn import metrics
Y_pred = model_dt.predict(X_test)
print(metrics.classification_report( y_true=Y_test, y_pred=Y_pred, target_names=labelencoder.classes_))
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20,20))
plot_tree(model_dt, feature_names=X_train.columns.values, class_names=labelencoder.classes_, rounded=True, filled=True)
plt.show()
```

# **Neural network ensamble**

```python
from sklearn.neural_network import MLPClassifier
model_rn = MLPClassifier(early_stopping=True,activation="relu",hidden_layer_sizes=(15), learning_rate='constant',
                     learning_rate_init=0.05, momentum= 0.035, max_iter=100, verbose=False)
model_rn.fit(X_train, Y_train)

Y_pred = model_rn.predict(X_test)
print(metrics.classification_report( y_true=Y_test, y_pred=Y_pred, target_names=labelencoder.classes_))
```

# **Pipe**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import FunctionTransformer, MinMaxScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle

data = pd.read_csv("Hotel Reservations.csv")
data['no_of_adults'] = data['no_of_adults'].astype(int)
data['no_of_children'] = data['no_of_children'].astype(int)
data['no_of_weekend_nights'] = data['no_of_weekend_nights'].astype(int)
data['no_of_week_nights'] = data['no_of_week_nights'].astype(int)
data['type_of_meal_plan'] = data['type_of_meal_plan'].astype(object)
data['required_car_parking_space'] = data['required_car_parking_space'].astype(int)
data['room_type_reserved'] = data['room_type_reserved'].astype(object)
data['lead_time'] = data['lead_time'].astype(int)
data['arrival_year'] = data['arrival_year'].astype(object)
data['arrival_month'] = data['arrival_month'].astype(int)
data['arrival_date'] = data['arrival_date'].astype(int)
data['market_segment_type'] = data['market_segment_type'].astype(object)
data['repeated_guest'] = data['repeated_guest'].astype(int)
data['no_of_previous_cancellations'] = data['no_of_previous_cancellations'].astype(int)
data['no_of_previous_bookings_not_canceled'] = data['no_of_previous_bookings_not_canceled'].astype(int)
data['avg_price_per_room'] = data['avg_price_per_room'].astype(float)
data['no_of_special_requests'] = data['no_of_special_requests'].astype(int)
data['booking_status'] = data['booking_status'].astype(object)
X = data.drop("booking_status", axis=1)
Y = data['booking_status']

categorical_cols = ['type_of_meal_plan','room_type_reserved','arrival_year','market_segment_type']
numeric_cols = ['no_of_adults','no_of_children','no_of_weekend_nights','no_of_week_nights','required_car_parking_space','lead_time','arrival_month','arrival_date','repeated_guest','no_of_previous_cancellations','no_of_previous_bookings_not_canceled','avg_price_per_room','no_of_special_requests']

num_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', MinMaxScaler())
])

cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('dummies', OneHotEncoder(drop='if_binary', handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', num_transformer, numeric_cols),
    ('cat', cat_transformer, categorical_cols)
])

pipe = Pipeline(steps=[
    ('preprocessor', preprocessor)
])

X_processed = pipe.fit_transform(X)

X_train, X_test, Y_train, Y_test = train_test_split(X_processed, Y, test_size=0.3)

from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(n_estimators=100, max_depth=15, random_state=100)

from sklearn.model_selection import GridSearchCV
criterion = ['gini', 'entropy']
min_samples_leaf = [10, 30, 50, 70, 100,200]
max_depth = [None, 10, 30, 50, 70, 100]
param_grid = dict(criterion=criterion, min_samples_leaf=min_samples_leaf, max_depth=max_depth)
grid = GridSearchCV(estimator=model_rf, param_grid=param_grid, scoring='f1_macro', n_jobs=-1, cv=3)
grid.fit(X_train, Y_train)
modelForest = grid.best_estimator_

!pip install scikit-optimize
from skopt.space import Real, Integer, Categorical
from skopt import BayesSearchCV
param_grid_rf = { 'max_depth': Integer(10, 100), 'min_samples_leaf': Integer(10, 100), 'n_estimators': Integer(50, 200), 'max_samples': Real(0.5, 0.8)}
bayes_search_rf = BayesSearchCV(estimator=modelForest, search_spaces=param_grid_rf, n_iter=50, cv=3, n_jobs=-1, scoring='accuracy', refit=True)
bayes_search_rf.fit(X_train, Y_train)
bayes_best_model_rf = bayes_search_rf.best_estimator_
best_params_rf = bayes_search_rf.best_params_

Y_pred = bayes_best_model_rf.predict(X_test)

medidas = pd.DataFrame()
accuracy = metrics.accuracy_score(Y_test, Y_pred)
precision = metrics.precision_score(Y_test, Y_pred, average='weighted')
recall = metrics.recall_score(Y_test, Y_pred, average='weighted')
f1 = metrics.f1_score(Y_test, Y_pred, average='weighted')
medidas['Forest'] = [accuracy, precision, recall, f1]
medidas.index = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
print(medidas)

pipeline_model = Pipeline([
    ('preprocessor', pipe),
    ('model', bayes_best_model_rf)
])
filename = 'pipeline_modelo_clasificacion.pkl'
pickle.dump(pipeline_model, open(filename, 'wb'))
```
